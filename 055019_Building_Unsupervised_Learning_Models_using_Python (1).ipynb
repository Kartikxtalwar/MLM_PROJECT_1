### **Analytical Report: Clustering of Import-Export Data**

**1. Introduction to Clustering**

In this analysis, we utilized K-Means Clustering, a widely adopted technique, to segment the dataset into distinct groups, known as clusters, based on their similarities. This method is particularly useful when there are no predefined labels, and the goal is to categorize the data based on inherent patterns. The primary objective of K-Means is to group data points in such a way that points within the same cluster are more similar to each other than to those in other clusters.

**2. Understanding the Dataset**

The dataset we analyzed contains information about various transactions related to imports and exports. The data includes multiple variables like:

* Transaction_ID: Unique ID for each transaction.
* Import_Export: Indicates whether the transaction is an import or export.
* Quantity, Value: Represents the quantity and monetary value involved in each transaction.
* Category: Type of product involved in the transaction.
* Weight: The weight of the items in the transaction.
* Shipping_Method, Supplier, Payment Terms: Information about how the transaction is processed, who the supplier is, and how payments are made.


**3. Selecting the Best Number of Clusters**

One of the challenges in K-Means clustering is determining the best number of groups (clusters) to use. To solve this:

* We created a Silhouette Plot: This plot helps in measuring how well each data point fits into its assigned cluster compared to other clusters.
* A higher Silhouette score suggests that the clusters are well-separated and the data points are well-grouped.

Based on this analysis, the optimal number of clusters was determined to be 2. This means that the data naturally divides into two distinct groups.

**4. Cluster Centers (Centroids)**

Each cluster is defined by a center, known as the centroid. This is a calculated average of all the data points within that cluster. The centroids for the two clusters are as follows:

* Cluster 0: The center of the first group, which represents data points with a certain set of characteristics.
* Cluster 1: The center of the second group, which represents a different set of characteristics.

The centroids show how the two clusters differ in terms of various features like the quantity, value, weight, and other parameters of the transactions.

3. Rescalling and Transformation

```python
# Identify numeric columns (non-categorical)
kk19_rs_num_col = kk19_rs.select_dtypes(include=[np.number]).columns.tolist()

# Rescale numeric columns using Min-Max Scaling
kk19_rs_scaler = MinMaxScaler()
kk19_rs[kk19_rs_num_col] = kk19_rs_scaler.fit_transform(kk19_rs[kk19_rs_num_col])

# Outlier Detection (Z-Score Method)
kk19_rs_outliers = {}
for kk19_rs_col in kk19_rs_num_col:
    z_scores = stats.zscore(kk19_rs[kk19_rs_col])
    kk19_rs_outliers[kk19_rs_col] = np.where(np.abs(z_scores) > 3)  # Identifying outliers with Z-score > 3

# Normality Testing using Jarque-Bera Test
kk19_rs_normality_results = {}
for kk19_rs_col in kk19_rs_num_col:
    jb_stat, jb_p_value = stats.jarque_bera(kk19_rs[kk19_rs_col])
    kk19_rs_normality_results[kk19_rs_col] = {'JB Statistic': jb_stat, 'P-value': jb_p_value}

# Prepare a Summary Table for the Analytical Report
summary_report = []

for kk19_rs_col in kk19_rs_num_col:
    # Outlier Detection Summary
    outliers = kk19_rs_outliers.get(kk19_rs_col, ([],))
    outlier_count = len(outliers[0])

    # Normality Test Summary
    jb_stat = kk19_rs_normality_results[kk19_rs_col]['JB Statistic']
    jb_p_value = kk19_rs_normality_results[kk19_rs_col]['P-value']

    # Rescaling Summary
    min_value = kk19_rs[kk19_rs_col].min()
    max_value = kk19_rs[kk19_rs_col].max()

    # Append results to summary report
    summary_report.append({
        'Variable': kk19_rs_col,
        'Outliers Detected': outlier_count,
        'JB Statistic': jb_stat,
        'JB P-value': jb_p_value,
        'Min After Rescaling': min_value,
        'Max After Rescaling': max_value
    })

# Convert the summary report to a DataFrame for better visualization
summary_report_df = pd.DataFrame(summary_report)

# Display the Summary Report Table
print(summary_report_df)

# Plot Outliers for Visualization (Boxplot)
for kk19_rs_col in kk19_rs_num_col:
    sns.boxplot(x=kk19_rs[kk19_rs_col])
    plt.title(f"Outliers in {kk19_rs_col}")
    plt.show()
```
